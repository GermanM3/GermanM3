```python
import pandas as pd
from pandas.tseries.offsets import MonthEnd, BDay
from datetime import timedelta
df_이익률 = pd.read_excel("그래프.xlsx", sheet_name="이익률")
df_현금흐름 = pd.read_excel("그래프.xlsx", sheet_name="현금흐름")
```

```python
# Filter rows to keep only the relevant financial data based on row index
이익률_data_indices = [8, 17, 26, 35]
현금흐름_data_indices = [8, 17, 26]


# Extract the relevant data rows and item names
이익률_data_rows = df_이익률.iloc[이익률_data_indices]
현금흐름_data_rows = df_현금흐름.iloc[현금흐름_data_indices]

# Extract the item names to be used as column headers
이익률_items = ['GP', 'Sales', 'OP', 'NP']
현금흐름_items = ['OCF', 'FCF', 'CAPEX']
all_items = 이익률_items + 현금흐름_items

# Select the data columns (from 'Unnamed: 3' onwards)
data_cols = [col for col in df_이익률.columns if col.startswith('Unnamed: ') and int(col.split(' ')[1]) >= 3]

# Extract the data values for the selected rows and columns
이익률_data_values = 이익률_data_rows[data_cols].values
현금흐름_data_values = 현금흐름_data_rows[data_cols].values

# Combine the data values using pd.concat
all_data_values = pd.concat([pd.DataFrame(이익률_data_values), pd.DataFrame(현금흐름_data_values)]).values

# Create the Period index from the original dataframes (Year is in row 5, Quarter is in row 4)
year_row = df_이익률.iloc[5, 3:]
quarter_row = df_이익률.iloc[4, 3:]
period_index = year_row.astype(str) + ' ' + quarter_row.astype(str)

# Create a new DataFrame with the correct structure, using all_items as column names
financial_data = pd.DataFrame(all_data_values.T, index=period_index.values, columns=all_items)

# Clean data values by removing commas and spaces and then convert to numeric
# Apply cleaning and conversion element-wise
financial_data = financial_data.apply(lambda col: pd.to_numeric(col.astype(str).str.replace(',', '', regex=False).str.replace(' ', '', regex=False), errors='coerce'))

# Clean up column names by removing '(천원)' and leading/trailing spaces
financial_data.columns = financial_data.columns.str.replace('(천원)', '', regex=False).str.strip()

```
